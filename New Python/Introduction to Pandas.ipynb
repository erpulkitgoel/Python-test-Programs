# Groupby 

```python
titanic_data.groupby('Pclass').describe()
```

```python
titanic_data.groupby('Sex').Age.min()
```

```python
gb = titanic_data.groupby('Sex')

def f(c):
    return np.mean(c)

gb.agg({'Age':['min','max'],'Fare':f})
```

```python
def gadd(r):
    return r['Sex'] + ':' + r['Name']

titanic_data['Gender-Name'] = titanic_data.apply(gadd, axis=1)
titanic_data.head()
```
