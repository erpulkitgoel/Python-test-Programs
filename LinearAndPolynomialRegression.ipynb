```python
from numpy import *

```

```python
x = array([0,1,2,3,4,5])
y = array([0,0.8,0.9,0.1,-0.8,-1])
print(x,y)
```

```python
from scipy.interpolate import *
```

```python
p1 = polyfit(x,y,1)
p2 = polyfit(x,y,2)
p3 = polyfit(x,y,3)
p4 = polyfit(x,y,4)
```

```python
print(p1)
print(p2)
print(p3)
print(p4)
```

```python
from matplotlib.pyplot import *
```

```python
plot(x,y,'o')
```

```python
plot(x,y,'o')
plot(x,polyval(p1,x),'r-')
plot(x,polyval(p2,x),'b-')
plot(x,polyval(p3,x),'g-')
plot(x,polyval(p4,x),'v-')
```

```python
plot(x,y,'o')
xp = linspace(-2,6,100)
plot(xp,polyval(p1,xp),'r-')
plot(xp,polyval(p2,xp),'b--')
plot(xp,polyval(p3,xp),'m:')

```

```python
# Lets predict the values of Y for each value of X based on linear regression
yfit = p1[0] * x + p1[1]
print(yfit)
print(y)
```

```python
yresid  = y - yfit
SSresid = sum(pow(yresid,2))
SStotal = len(y) * var(y)
rsd = 1 - SSresid/SStotal
print(rsd)
```

```python
from scipy.stats import *
slope, intercept, r_value, p_value, std_err = linregress(x,y)
print(pow(r_value,2))
```
